# College Completion Rates EDA

## Analysis

Questions/Hypotheses:
Are there connections between tuition costs/types of institution, and completion rates? Do loans/grants matter? I hypothesize that institutions with higher tuition have higher completion rates because I think that higher income families have more students that go to institutions with higher tuition, and these more fortunate students have a better chance of completing college. As such, I think it would make sense that private non-profit schools have the highest completion rates because they usually cost more. Along the same train of thought, I feel that institutions with lower percentages of loans/grants will have higher completion rates because they require less aid, and thus may be higher income families as well.

Findings/Correlations:
For this exploration, I chose to bin my cost, loan %, and grant % data into 3 bins: low, medium, and high. When exploring scatterplots of the data initially, I found that the scatter plots were a big mess and couldn't tell me much. So, I figured I could get more information by grouping these columns into categories, and plotting bar charts. 
I found that my first two hypotheses were correct. As seen in Figure 1.1, schools with higher cost ("high" bin) have the highest completion percentages, and in Figure 1.2, private non-profit schools have the highest completion percentages. However, when performing bivariate analysis with the two, I found that the combination of high cost schools and public schools produced the highest completion percentage.
When exploring loans/grants, I found that my hypothesis about grants was correct, but my hypothesis about loans wasn't. As seen in Figure 2.1, there doesn't seem to be too much connection between loans and completion percentages, but there is a negative relationship seen with grants, as schools with lower grant percentages have higher completion rates. I believe there isn't much connection between loans and completion percentages because loans aren't a good indicator of family income, since grants are usually only given to lower income families whereas almost everyone can take loans, and there is a higher mean loan percentage amongst colleges than grant percentages. 

Shortcomings:
My biggest issue with this exploration was the data shortage. I found that many schools do not report completion percentages, costs, and loan/grant information. My initial dataset was about 6000 rows in size, but after cleaning and extracting the information I wanted, I was down to about 2000 rows. Using about 1/3 of the data available is less than ideal, but looking at Figures 0.1 and 0.2, the spread across the US seemed about the same, so the subset dataset is still somewhat representative of the original dataset. 

```python
import numpy as np
import pandas as pd
import os
%matplotlib inline
import project02 as proj
```

```python
college_path = os.path.join('data', 'MERGED2016_17_PP.csv')
college = pd.read_csv(college_path)
```

```python
datadict_path = os.path.join('data', 'CollegeScorecardDataDictionary.xlsx')
datadict = pd.read_excel(datadict_path, sheet_name='data_dictionary')
```

# Data Cleaning
Creating, cleaning, and preparing a dataset for data exploration

```python
# Pick wanted columns
aid = datadict[datadict['dev-category'] == 'aid']['VARIABLE NAME'] # columns regarding financial aid
cost = datadict[datadict['dev-category'] == 'cost']['VARIABLE NAME'] # columns regarding cost
cols = pd.Series(college.columns)
wants = ['CONTROL', 'PREDDEG', 'C150_4_POOLED_SUPP'] # other wanted columns

# Construct DF
df = college[cols[cols.isin(aid) | cols.isin(cost) | cols.isin(wants)]]
avgcost = df[cols[cols.str[:3] == 'NPT']].apply(lambda x: np.mean(x), axis=1)
df = df.drop(cols[cols.str[:3] == 'NPT'], axis=1)
df['AVERAGE COST'] = avgcost
df = df.rename(columns={'C150_4_POOLED_SUPP': 'COMPLETION PERCENTAGE'})
df = df.drop(['GRAD_DEBT_MDN_SUPP', 'GRAD_DEBT_MDN10YR_SUPP'], axis=1)
df = df.dropna().loc[~(df == 'PrivacySuppressed').any(axis=1)]
subset = df.copy()
df = df.apply(pd.to_numeric).reset_index(drop=True)
```

# Feature Engineering
Creating new features that will be more useful for grouping and data visualization

```python
# Visualizing histogram distributions of cost, loans, and grants
df['AVERAGE COST'].plot.hist(bins=12).set_title('Average Cost Histogram')
```

```python
df['PCTFLOAN'].plot.hist(bins=12).set_title('Percent Loans Histogram')
```

```python
df['PCTPELL'].plot.hist(bins=12).set_title('Percent Pell Grant Histogram')
```

```python
# categorizing cost, loans, and grants, and adding them as new columns
def categorize(val):
    '''
    Categorizes the bins as low, med, or high:
    First four bins = low
    Middle four bins = med
    Last four bins = high
    '''
    if 1 <= val <= 4:
        return 'low'
    if 5 <= val <= 8:
        return 'med'
    return 'high'

costs = pd.qcut(df['AVERAGE COST'], 12, labels = np.arange(1, 13)).apply(categorize)
loans = pd.qcut(df['PCTFLOAN'], 12, labels = np.arange(1, 13)).apply(categorize)
pells = pd.qcut(df['PCTPELL'], 12, labels= np.arange(1, 13)).apply(categorize)
df['COST'] = costs
df['LOAN'] = loans
df['PELL'] = pells
df = df.replace(proj.translation_dict(datadict))

# new df
df.head()
```

```python
# data on the new columns
cost_means = df.groupby('COST').agg({'AVERAGE COST': ['min', 'max', 'mean']})
loan_means = df.groupby('LOAN').agg({'PCTFLOAN': ['min', 'max', 'mean']})
pell_means = df.groupby('PELL').agg({'PCTPELL': ['min', 'max', 'mean']})
pd.concat([cost_means, loan_means, pell_means], axis=1).sort_values(by=('AVERAGE COST', 'min'))
```

# Cost/Control Exploration
Exploring connections between loans/grants and college completion percentage

```python
# Cost-Completion Percentage Univariate
cost = df.copy()
cost = cost.groupby('COST').agg({'COMPLETION PERCENTAGE': ['mean', 'median']})
cost.columns = [colu[1] for colu in cost.columns]
cost = cost.sort_values(by='mean')
display(cost)
cost.plot.bar().set_title('Figure 1.1: Mean/Median Completion Percentage of Cost Bins')
```

```python
# Control-Completion Percentage Univariate
ctrl = df.copy()
ctrl = ctrl.groupby(['CONTROL']).agg({'COMPLETION PERCENTAGE': ['mean', 'median']})
ctrl.columns = [colu[1] for colu in ctrl.columns]
ctrl = ctrl.sort_values(by='mean')
display(ctrl)
ctrl.plot.bar().set_title('Figure 1.2: Mean/Median Completion Percentage of Control Types')
```

```python
# Bivariate (cost/control-completion percentage)
biv = df.copy()
biv = biv.groupby(['CONTROL', 'COST']).agg({'COMPLETION PERCENTAGE': ['mean', 'median']})
biv.columns = [colu[1] for colu in biv.columns]
biv = biv.sort_values(by='mean')
display(biv)
biv.plot.bar().set_title('Figure 1.3: Mean/Median Completion Percentage of Cost Bins and Control Types')
```

# Loans/Grants exploration
Exploring connections between loans/grants and college completion percentage

```python
# Loan Univariate
loan = df.copy()
loan = loan.groupby(['LOAN']).agg({'COMPLETION PERCENTAGE': ['mean', 'median']})
loan.columns = [colu[1] for colu in loan.columns]
loan = loan.sort_values(by='mean')
display(loan)
loan.plot.bar().set_title('Figure 2.1: Mean/Median Completion Percentage of Loan Bins')
```

```python
# Pell Univariate
pell = df.copy()
pell = pell.groupby(['PELL']).agg({'COMPLETION PERCENTAGE': ['mean', 'median']})
pell.columns = [colu[1] for colu in pell.columns]
pell = pell.sort_values(by='mean')
display(pell)
pell.plot.bar().set_title('Figure 2.2: Mean/Median Completion Percentage of Pell Grant Bins')
```
